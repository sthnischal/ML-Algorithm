```python
#visualize correlation table in the form of a heatmap for better understanding and presentation
#Pairwise correlation heatmap

import seaborn as sns
ax = sns.heatmap(dataset.corr(),
                cmap = sns.cubehelix_palette(20, light=0.95, dark=0.15))
ax.xaxis.tick_top() #move the labels to the top

plt.savefig('./housing-coorr.png', 
            bbox_inches='tight', dpi=300)
```

```python
#use seaborn pairplot function to visualize the data using the pairplot graph
sns.pairplot(dataset,
            plot_kws={'alpha': 0.6},
             diag_kws={'bins': 30})
```

```python
#This shows 
    #RM and MEDV have the shape like that in a normally distributed graph.
    #AGE is skewed to the left and LSTAT is skewed to the right.
    #TAX has a large amount of distribution around the point 700.
```

```python
#RM and LSTAT feature are compare with target variable MEDV because they are highly correlated
#Modeling MADV as a function of RM and LSTAT
fig, ax = plt.subplots(1,2)
sns.regplot('RM', 'MEDV', dataset, ax=ax[0], scatter_kws={'alpha': 0/4})
sns.regplot('LSTAT', 'MEDV',dataset, ax=ax[1], scatter_kws={'alpha': 0.4})
plt.savefig('./housing-scatter.png',
           bbox_inches='tight', dpi=300)
```
